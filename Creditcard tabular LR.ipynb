```python
#undersampling method
#splitting dataset into independant and dependent feautres
X=data.drop(['Class'], axis = 1)
X.head()
X.shape

Y=data['Class']
Y.head()
Y.shape
```

```python
#undersampling
from imblearn import under_sampling
from imblearn.under_sampling import NearMiss
```

```python
'''#implementing nearmiss to handle imbalanced dataset
nm = NearMiss()
X_1 ,Y_1 = nm.fit_sample(X,Y)
X_1.shape
Y_1.shape'''


# Assuming you have your features in X and labels in Y
nm = NearMiss()
X_1, Y_1 = nm.fit_resample(X, Y)

print(X_1.shape)
print(Y_1.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train, y_test = train_test_split(X_1, Y_1,test_size= 0.20,random_state=40)
print("x_train :" , x_train.shape)
print("x_test :" , x_test.shape)
print("y_train :" , y_train.shape)
print("y_test :" , y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cm = confusion_matrix(y_test ,y_predicted)
print("Confusion_matrix : ")
print(cm)

print("Accuracy_score :", accuracy_score (y_test, y_predicted))
print(classification_report(y_test, y_predicted))
```
